# EDA on Vehicle Insurance Customer Data

```python
import pandas as pa
```

```python
customer_details=pa.read_csv('.\data\customer_details.csv')
customer_details.columns=['customer_id','Gender','age','driving licence present','region code','previously insured','vehicle age','vehicle damage,in respective order']
```

```python
customer_details
```

```python
customer_policy=pa.read_csv('.\data\customer_policy_details.csv')
customer_policy.columns=['customer_id','annual premium (in Rs)',"sales channel code",'vintage','response']
customer_policy
```

## 2.Checking and cleaning Data Quality

```python
d1=customer_details.isnull().sum()
print('The count of the null values in columns of customer_details')
print('*'*60)
print('Columns\t\t\t\t   Null values')
print('*'*60)
print(d1)
```

```python
d2=customer_policy.isnull().sum()
print('The count of the null values in columns of customer_policy')
print('*'*60)
print('Columns\t\t\tNull values')
print('*'*60)
print(d2)
```

```python
#Drop Null
customer_policy1=customer_policy.dropna(subset =['customer_id'])
```

```python
customer_details1=customer_details.dropna(subset =['customer_id'])
```

```python
customer_details1.isnull().sum()
```

```python
customer_policy1.isnull().sum()
```

```python
# Replace null values
#mean
customer_details1['age']=customer_details1['age'].fillna(customer_details1['age'].mean())
customer_details1['region code']=customer_details1['region code'].fillna(customer_details1['region code'].mean())

#mode
customer_details1['vehicle age']=customer_details1['vehicle age'].fillna(customer_details1['vehicle age'].mode()[0])
customer_details1['Gender']=customer_details1['Gender'].fillna(customer_details1['Gender'].mode()[0])
customer_details1['driving licence present']=customer_details1['driving licence present'].fillna(customer_details1['driving licence present'].mode()[0])
customer_details1['previously insured']=customer_details1['previously insured'].fillna(customer_details1['previously insured'].mode()[0])
customer_details1['vehicle damage,in respective order']=customer_details1['vehicle damage,in respective order'].fillna(customer_details1['vehicle damage,in respective order'].mode()[0])
```

```python
customer_details1.isnull().sum()
```

```python
customer_details1
```

```python
#mean
customer_policy1['annual premium (in Rs)']=customer_policy1['annual premium (in Rs)'].fillna(customer_policy1['annual premium (in Rs)'].mean())
customer_policy1['sales channel code']=customer_policy1['sales channel code'].fillna(customer_policy1['sales channel code'].mean())
customer_policy1['vintage']=customer_policy1['vintage'].fillna(customer_policy1['vintage'].mean())
#mode
customer_policy1['response']=customer_policy1['response'].fillna(customer_policy1['response'].mode()[0])
```

```python
# Remove White Space
customer_details1['Gender']=customer_details1['Gender'].str.strip()
customer_details1['vehicle age']=customer_details1['vehicle age'].str.strip()
```

```python
# case correction
customer_details1['Gender']=customer_details1['Gender'].str.upper()
customer_details1['vehicle age']=customer_details1['vehicle age'].str.upper()
```

```python
#Dummies
cf1=pa.get_dummies(customer_details1['previously insured'])
cf2=pa.get_dummies(customer_details1['driving licence present'])
cf3=pa.get_dummies(customer_details1['vehicle damage,in respective order'])
cf4=pa.get_dummies(customer_details1['Gender'])
print(cf1)
print(cf2)
print(cf3)
print(cf4)
```

## 3.Create Master Table

```python
master_table=pa.merge(customer_details1,customer_policy1)
master_table
```

## 4.Info

```python
# Gender wise Avg annual premium
print('Gender wise Avg annual premium\n')
gic=master_table.groupby(['Gender']).mean()['annual premium (in Rs)']
print(gic)
```

```python
# Age wise Avg annual premium
print('Age wise Avg annual premium\n')
aic=master_table.groupby(['age']).mean()['annual premium (in Rs)']
print(aic)
```

```python
#Data balance between Genders
gbal=master_table.groupby(['Gender']).sum()
print(gbal)
```

```python
# Vehicle Age wise Avg annual premium
print('Vehicle Age wise Avg annual premium\n')
vic=master_table.groupby(['vehicle age']).mean()['annual premium (in Rs)']
print(vic)
```
